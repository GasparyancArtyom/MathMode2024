```python
def method1(function, start, stop, rise):
    return rise * sum(function(np.arange(start, stop, rise))) 
```

```python
def method2(function, start, stop, rise):
    return (rise/2) * (function(start) + 2*sum(function(np.arange(start, stop, rise)[1:-1])) + function(stop))
```

```python
def method3(function, start, stop, rise):
    interval = np.arange(start, stop, rise)[1:-1]
    return (function(start) + 2*sum([function(interval[i]) for i in range(len(interval)) if i % 2 == 0]) + 4*sum([function(interval[i]) for i in range(len(interval)) if i % 2 == 1]) + function(stop)) * (rise/3) 
```

```python
step = 0.1
ERR = 10e7
while ERR > 1e-5:
     METHOD1 = method1(lambda x: np.exp(1/(x**2))/(x**3), 1, 2, step)
     METHOD2 = method2(lambda x: np.exp(1/(x**2))/(x**3), 1, 2, step)
     METHOD3 = method3(lambda x: np.exp(1/(x**2))/(x**3), 1, 2, step)
     ERR = max([abs(METHOD1-METHOD2), abs(METHOD1-METHOD3), abs(METHOD2-METHOD3)])
     step *= 0.1
     print(METHOD1, METHOD2, METHOD3)
```
